## 3. Explain ‚Äî De la br√∫jula f√≠sica al estado cu√°ntico ‚à£œà‚ü© = Œ±‚à£0‚ü© + Œ≤‚à£1‚ü©

**Objetivo:**  
Formalizar la experiencia f√≠sica con la tirita y proyectarla al plano matem√°tico: modelado del vector estado tridimensional.. 

**Paso a paso (docente gu√≠a)**

### 3.1 De los polos al eje Z

En el plano vertical del tablero, identifiquemos dos estados extremos:

- Cuando la aguja apunta totalmente hacia el **norte** (hacia arriba del eje Z):  
  ‚Üí lo llamaremos estado **‚à£0‚ü©**.
- Cuando apunta completamente hacia el **sur** (hacia abajo del eje Z):  
  ‚Üí ser√° el estado **‚à£1‚ü©**.

Estos dos estados forman nuestra **base computacional**, es decir, los puntos de referencia fundamentales.

### 3.2 Una br√∫jula entre dos mundos

Si inclinamos la aguja, ya no est√° totalmente en ‚à£0‚ü© ni en ‚à£1‚ü©.  
Podemos describir su posici√≥n intermedia con un **√°ngulo Œ∏**, medido desde el eje Z.  
La proyecci√≥n sobre **Z+** corresponde a la **probabilidad de estar en ‚à£0‚ü©**, y la proyecci√≥n sobre **Z‚àí** a la **probabilidad de estar en ‚à£1‚ü©**.

De esta forma:

$$
P(0) = \cos^2\left(\frac{Œ∏}{2}\right), \qquad
P(1) = \sin^2\left(\frac{Œ∏}{2}\right)
$$

Estas probabilidades siempre suman 1.

### 3.3 De √°ngulos a amplitudes complejas

La orientaci√≥n completa de la br√∫jula no depende solo del √°ngulo Œ∏, sino tambi√©n del giro alrededor del eje Z, que llamamos **fase œÜ**.  
Ese giro no cambia la inclinaci√≥n, pero s√≠ la ‚Äúrotaci√≥n interna‚Äù del vector.

Combinando ambos par√°metros, podemos escribir el estado cu√°ntico como:

$$
‚à£œà‚ü© = \cos\left(\frac{Œ∏}{2}\right)‚à£0‚ü© + e^{iœÜ}\sin\left(\frac{Œ∏}{2}\right)‚à£1‚ü©
$$

Aqu√≠:
- $Œ± = \cos\left(\frac{Œ∏}{2}\right)$  
- $Œ≤ = e^{iœÜ}\sin\left(\frac{Œ∏}{2}\right)$  

Y cumplen la condici√≥n fundamental:  
$$|Œ±|^2 + |Œ≤|^2 = 1.$$

## 4 Elaborate ‚Äî Simulando la br√∫jula cu√°ntica en Qiskit

Ahora traduciremos la orientaci√≥n de nuestra br√∫jula tridimensional al **lenguaje cu√°ntico**.  
Cada √°ngulo $(Œ∏, œÜ)$ representa un estado puro en la esfera de Bloch.  
El objetivo es **visualizar el estado ‚à£œà‚ü©** definido por esos √°ngulos y entender c√≥mo las rotaciones cambian su posici√≥n.



### 4.1 Definici√≥n del estado cu√°ntico

Comenzaremos creando un circuito de un solo qubit.  
Aplicaremos rotaciones para representar los √°ngulos Œ∏ (inclinaci√≥n) y œÜ (fase).

```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_bloch_multivector
from qiskit.quantum_info import Statevector
import numpy as np

# Par√°metros de orientaci√≥n
theta = np.pi / 3   # Inclinaci√≥n del vector (60¬∞)
phi = np.pi / 4     # Fase o giro alrededor del eje Z (45¬∞)

# Circuito de un qubit
qc = QuantumCircuit(1)
qc.ry(theta, 0)     # Rotaci√≥n en Y -> controla el √°ngulo Œ∏
qc.rz(phi, 0)       # Rotaci√≥n en Z -> controla la fase œÜ
qc.draw('mpl')

# Obtener el vector de estado
state = Statevector.from_instruction(qc)
display(plot_bloch_multivector(state))


### 4.2Ô∏è Exploraci√≥n guiada

> **Demostraci√≥n docente para el docente:**  
> - Modifica los valores de `theta` y `phi` en el c√≥digo anterior para observar c√≥mo cambia la orientaci√≥n del vector en la esfera de Bloch.  
> - Prop√≥n las siguientes preguntas a los docentes en formaci√≥n:  
>   - ¬øQu√© sucede cuando $Œ∏ = 0$ o $Œ∏ = œÄ$?  
>     *(Pista: el vector apunta al polo norte o sur de la esfera, representando ‚à£0‚ü© o ‚à£1‚ü©).*  
>   - ¬øQu√© pasa si cambiamos œÜ manteniendo Œ∏ fijo?  
>     *(Pista: el vector gira alrededor del eje Z sin alterar las probabilidades de medici√≥n, solo su fase relativa).*  
> - Conecta visualmente este comportamiento con la br√∫jula tridimensional del ejercicio anterior:  
>   la rotaci√≥n en **Y** inclina la aguja, y la rotaci√≥n en **Z** la hace girar alrededor de su eje.

> üí¨ **Discusi√≥n guiada:**  
> ‚ÄúSi la br√∫jula cu√°ntica se mueve libremente sobre la esfera, ¬øqu√© representa el color o direcci√≥n de su sombra en el plano XY?‚Äù  
> Este di√°logo busca que los docentes reconozcan la **fase cu√°ntica** como una propiedad que no cambia la probabilidad, pero s√≠ la **coherencia** del estado.



### 4.3Ô∏è Conexi√≥n conceptual

La simulaci√≥n en Qiskit ilustra que un **estado cu√°ntico puro** puede escribirse como:  

$$
‚à£œà‚ü© = \cos\left(\frac{Œ∏}{2}\right)‚à£0‚ü© + e^{iœÜ}\sin\left(\frac{Œ∏}{2}\right)‚à£1‚ü©
$$  

donde:
- $Œ∏$ controla la **inclinaci√≥n** del vector (an√°loga al √°ngulo de la br√∫jula).
- $œÜ$ representa la **fase** o rotaci√≥n en torno al eje Z.  

> **Interpretaci√≥n pedag√≥gica:**  
> - $|Œ±|^2 = \cos^2\left(\frac{Œ∏}{2}\right)$ ‚Üí probabilidad de encontrar el qubit en el estado ‚à£0‚ü©.  
> - $|Œ≤|^2 = \sin^2\left(\frac{Œ∏}{2}\right)$ ‚Üí probabilidad de encontrarlo en ‚à£1‚ü©.  

> **Sugerencia para el docente:**  
> Dibuja en el tablero una semiesfera de Bloch con cinta o hilo, y muestra c√≥mo un cambio en $Œ∏$ y $œÜ$ ‚Äúmueve‚Äù el vector ‚à£œà‚ü©.  
> Esto ayuda a que los participantes visualicen la conexi√≥n entre **rotaciones f√≠sicas** y **transformaciones unitarias**.


## 5 Evaluate ‚Äî Reflexi√≥n y evaluaci√≥n formativa

> **Actividad de cierre: ‚ÄúIdentifica tu estado cu√°ntico favorito‚Äù**

Cada docente selecciona un par de √°ngulos $(Œ∏, œÜ)$ que representen **su propia br√∫jula cu√°ntica**.  
Luego, los ingresan en el c√≥digo para visualizar en la esfera de Bloch la direcci√≥n que toma su ‚à£œà‚ü©.

**Preguntas de reflexi√≥n guiada:**
- ¬øC√≥mo se interpreta el cambio en $Œ∏$ y $œÜ$ en t√©rminos de probabilidades de medici√≥n?
- ¬øQu√© aspecto del movimiento te record√≥ m√°s al comportamiento de una br√∫jula f√≠sica?
- ¬øPor qu√© se dice que cambiar la fase no altera el resultado de una medici√≥n, pero s√≠ la coherencia del estado?

> **Sugerencia para el docente:**  
> Invita a los participantes a compartir en grupo sus visualizaciones (capturas o diagramas dibujados) y a explicar brevemente **c√≥mo su elecci√≥n de √°ngulos representa un estado posible del sistema**.  
> Esto fomenta la comprensi√≥n simb√≥lica y visual de los estados cu√°nticos.




---

>  *(Espacio reservado para QR o enlace de video del docente sobre la semiesfera de Bloch y su proyecci√≥n visual.)*
