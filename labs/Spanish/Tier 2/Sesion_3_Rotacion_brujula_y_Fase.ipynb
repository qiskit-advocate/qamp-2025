## 2. Explore — Actividad manipulativa: Tira de papel y proyección de fase

**Materiales:**  
Tira de papel (10–15 cm), cinta adhesiva, linterna o foco, pizarra blanca o pared.

**Procedimiento:**  
1. Pega la tirita verticalmente en la pared o tablero.  
2. Ilumínala con la linterna perpendicularmente: observa la sombra.  
3. Gira la tirita unos grados → la sombra cambia de posición.  
4. Mueve la linterna (sin rotar la tirita) → la sombra cambia aunque la orientación física sea la misma.  

**Propósito:**  
- Construir una intuición física de la “fase”: la orientación real es la tirita, las sombras son las *componentes observables*.  
- Introducir la idea de una **segunda dimensión de información** no visible directamente (fase).

## 3. Explain — De la observación al modelo matemático

**Objetivo:**  
Formalizar la experiencia física con la tirita y proyectarla al plano matemático: cómo una rotación o una fase afectan al estado, sin alterar su norma.

**Paso a paso (docente guía)**

1. **La tirita como vector**  
   - La orientación real de la tirita representa el vector magnético o el estado de un sistema:  
     $ vec{M}(\theta) = (\cos\theta, \sin\theta) $  
   - Su sombra en el eje X es $\cos\theta$ y en Y es $\sin\theta$.  

2. **Rotación física**  
   - Girar la tirita cambia ambas componentes: la sombra “camina” de un eje a otro.  
   - Este cambio puede representarse con una matriz de rotación:  
     $$
     R(\theta) =
     \begin{pmatrix}
     \cos\theta & -\sin\theta \\
     \sin\theta & \cos\theta
     \end{pmatrix}
     $$

3. **Fase — La dimensión invisible**  
   - Si cambias la posición de la linterna (sin girar la tirita), la sombra cambia aunque el objeto no lo haga.  
   - Matemáticamente, eso equivale a multiplicar por una **fase compleja** $e^{i\phi}$:  
     $$
     z' = e^{i\phi} z, \quad z = \cos\theta + i\sin\theta
     $$  
   - La fase no cambia la magnitud (‖z‖ = 1), solo su orientación interna.

4. **Interpretación didáctica**  
   - La tirita → estado completo del sistema.  
   - La sombra → lo que se mide (proyección).  
   - La linterna → el observador o base de medida.  

5. **Cierre conceptual**  
   - Rotar el sistema modifica sus componentes observables.  
   - Cambiar la fase modifica cómo se perciben las componentes sin alterar el estado físico.  
   - En un qubit, estas operaciones son análogas a rotaciones en la esfera de Bloch.

## 4. Elaborate — El código como simulador de fases


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
theta = np.linspace(0, 2*np.pi, 200)
phi = np.pi / 4  # fase adicional

# Estado base y con fase
z_base = np.exp(1j * theta)
z_phase = np.exp(1j * (theta + phi))

plt.figure(figsize=(6, 6))
plt.plot(np.real(z_base), np.imag(z_base), label='Estado base')
plt.plot(np.real(z_phase), np.imag(z_phase), '--', label='Con fase φ=π/4')
plt.xlabel('Re')
plt.ylabel('Im')
plt.title('Rotación del estado y cambio de fase')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```

**Interpretación del código**
- Cada punto representa un estado posible del sistema (unidad en el plano complejo).  
- La rotación por $\theta$ representa el cambio físico de orientación.  
- La fase $\phi$ desplaza todo el estado en la misma dirección sin alterar su módulo.
