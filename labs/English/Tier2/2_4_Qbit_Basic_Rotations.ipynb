````python
# Qiskit code â€” Rotation and Bloch Visualization
from qiskit import QuantumCircuit
from qiskit.visualization import plot_bloch_multivector
from qiskit.quantum_info import Statevector
import numpy as np

# Example angles
theta_x = np.pi / 2
theta_y = np.pi / 3
theta_z = np.pi / 4

# Create a single qubit circuit
qc = QuantumCircuit(1)

# Apply rotations
qc.rx(theta_x, 0)
qc.ry(theta_y, 0)
qc.rz(theta_z, 0)

# Get the final statevector
state = Statevector.from_instruction(qc)

# Display circuit and Bloch vector
display(qc.draw('mpl'))
plot_bloch_multivector(state)
````
