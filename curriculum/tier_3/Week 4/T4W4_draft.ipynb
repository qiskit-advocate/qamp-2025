**Objective:** Provide a formal, step-by-step mathematical derivation of the quantum teleportation protocol. This mini-project will synthesize all Tier 3 concepts covered so far (state vectors, unitary gates, tensor products, and entanglement).

## 1. Bridging from Past Weeks

Quantum Teleportation is the perfect summaries for our study of single and two-qubit systems. It's a protocol that *uses* all the concepts we've mastered:

1.  **Week 1 (State Vectors):** The entire goal is to transmit an arbitrary, unknown quantum state (a state vector) $|\psi
\rangle = \alpha|0
\rangle + \beta|1
\rangle$ from one location to another.
2.  **Week 3 (Entanglement & Tensor Products):** The protocol is *impossible* without a pre-shared entangled Bell state, $|\Phi^+
angle$. The entire calculation will be done by tracking the tensor product of the 3-qubit system.
3.  **Week 2 (Unitary Gates):** The final step of the protocol relies on Bob applying conditional $X$ and $Z$ gates (which are just $R_x(\pi)$ and $R_z(\pi)$ rotations) to recover the original state.

**What Teleportation IS and ISN'T:**
* **IS:** A way to transmit a quantum *state* from one place to another, destroying the original in the process (due to the No-Cloning Theorem, for more info can refer Tier 4 Week 1).
* **ISN'T:** Instantaneous communication. The protocol requires sending two classical bits of information. This classical communication is limited by the speed of light, so the *full* state transfer is not faster-than-light (FTL).

## 2. The Setup & Initial State

We have a 3-qubit system. Let's label the qubits:
* **$q_C$ (Cat):** The unknown state $|\psi
\rangle$ that Alice wants to send. 
<br>(*REMEMBER* that the state that are going to teleport is an information, not a physical cat to be transport)<br>
$|\psi\rangle_C = \alpha|0\rangle + \beta|1\rangle$.
* **$q_A$ (Alice):** Alice's half of a pre-shared Bell pair.
* **$q_B$ (Bob):** Bob's half of the pre-shared Bell pair.

Alice holds $q_C$ and $q_A$. Bob holds $q_B$ and is far away.

The pre-shared Bell state is $|q_A q_B\rangle = |\Phi^+\rangle_{AB} = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle)$.

The total initial state of the 3-qubit system, $|\Psi_{total}\rangle$, is the tensor product of Alice's cargo qubit and the Bell pair:

$$ 
|\Psi_{total}
\rangle = |\psi
\rangle_C \otimes |\Phi^+
\rangle_{AB} 
$$ 

Let's expand this (qubit order will be $C, A, B$):

$$ 
|\Psi_{total}
\rangle = (\alpha|0
\rangle_C + \beta|1
\rangle_C) \otimes \frac{1}{\sqrt{2}}(|00
\rangle_{AB} + |11
\rangle_{AB}) 
$$ 

$$ 
|\Psi_{total}
\rangle = \frac{1}{\sqrt{2}} \left[ \alpha|0
\rangle_C \otimes (|00
\rangle_{AB} + |11
\rangle_{AB}) + \beta|1
\rangle_C \otimes (|00
\rangle_{AB} + |11
\rangle_{AB}) 
\right] 
$$ 

$$ 
|\Psi_{total}
\rangle = \frac{1}{\sqrt{2}} \left[ \alpha(|000
\rangle + |011
\rangle) + \beta(|100
\rangle + |111
\rangle) 
\right] 
$$ 

This is our starting point. Alice has the first two qubits; Bob has the third.

--- 

## 3. Alice's Operations: The Bell Basis Measurement

Alice can't just measure $|\psi
\rangle$—that would destroy it. Instead, she performs a "Bell Basis Measurement" on her two qubits ($q_C$ and $q_A$). This involves two gates:

1.  **$CNOT$ gate:** Alice applies a $CNOT$ with $q_C$ as the control and $q_A$ as the target.
2.  **$Hadamard$ gate:** Alice applies an $H$ gate to $q_C$.

Let's trace the state vector through these two operations.

### Step A: After Alice's $CNOT$ (control: $q_C$, target: $q_A$)

The $CNOT$ flips $q_A$ if $q_C$ is 1. We apply this to our four terms:
* $\alpha|000
\rangle \to \alpha|000
\rangle$ (No change)
* $\alpha|011
\rangle \to \alpha|011
\rangle$ (No change)
* $\beta|100
\rangle \to \beta|110
\rangle$ ($q_A$ flips from 0 to 1)
* $\beta|111
\rangle \to \beta|101
\rangle$ ($q_A$ flips from 1 to 0)

The state is now $|\Psi_1
\rangle$:
$$ 
|\Psi_1
\rangle = \frac{1}{\sqrt{2}} \left[ \alpha(|000
\rangle + |011
\rangle) + \beta(|110
\rangle + |101
\rangle) 
\right] 
$$ 

### Step B: After Alice's $Hadamard$ (on $q_C$)

Now Alice applies $H$ to $q_C$ (the first qubit).
* $|0
\rangle \to H|0
\rangle = \frac{1}{\sqrt{2}}(|0
\rangle + |1
\rangle)$
* $|1
\rangle \to H|1
\rangle = \frac{1}{\sqrt{2}}(|0
\rangle - |1
\rangle)$

We apply this to all four terms in $|\Psi_1
\rangle$:
$$ 
|\Psi_2
\rangle = \frac{1}{\sqrt{2}} \left[ \alpha\left(\frac{1}{\sqrt{2}}(|0
\rangle + |1
\rangle)|00
\rangle + \frac{1}{\sqrt{2}}(|0
\rangle + |1
\rangle)|11
\rangle
\right) + \beta\left(\frac{1}{\sqrt{2}}(|0
\rangle - |1
\rangle)|10
\rangle + \frac{1}{\sqrt{2}}(|0
\rangle - |1
\rangle)|01
\rangle
\right) 
\right] 
$$ 

Let's expand this and pull the $\frac{1}{\sqrt{2}}$ out front (total pre-factor is $\frac{1}{2}$):
$$ 
|\Psi_2
\rangle = \frac{1}{2} \left[ \alpha(|000
\rangle + |100
\rangle + |011
\rangle + |111
\rangle) + \beta(|010
\rangle - |110
\rangle + |001
\rangle - |101
\rangle) 
\right] 
$$

--- 

## 4. The "Magic": Regrouping by Alice's Basis States

This is the most critical step of the derivation. The state $|\Psi_2
\rangle$ looks like a mess. But let's *regroup* the terms based on what Alice controls ($q_C, q_A$) to see what has happened to Bob's qubit ($q_B$).

We will factor out Alice's four possible basis states: $|00
\rangle_{CA}$, $|01
\rangle_{CA}$, $|10
\rangle_{CA}$, $|11
\rangle_{CA}$.

$$ 
|\Psi_2\rangle = \frac{1}{2} \left[ \alpha|000\rangle+\beta|001\rangle+
\alpha|011\rangle+\beta|010\rangle+
\alpha|100\rangle-\beta|101\rangle+
\alpha|111\rangle-\beta|110\rangle]\right.
$$ 

* **$|00
\rangle_{CA}$:** (Find all terms that start with $|00
\rangle$)
    * $\alpha|000
\rangle \to \alpha|0
\rangle_B$
    * $\beta|001
\rangle \to \beta|1
\rangle_B$
    * $\implies |00
\rangle_{CA} \otimes (\alpha|0
\rangle + \beta|1
\rangle)_B$

* **$|01
\rangle_{CA}$:** (Find all terms that start with $|01
\rangle$)
    * $\alpha|011
\rangle \to \alpha|1
\rangle_B$
    * $\beta|010
\rangle \to \beta|0
\rangle_B$
    * $\implies |01
\rangle_{CA} \otimes (\beta|0
\rangle + \alpha|1
\rangle)_B$

* **$|10
\rangle_{CA}$:** (Find all terms that start with $|10
\rangle$)
    * $\alpha|100
\rangle \to \alpha|0
\rangle_B$
    * $\beta(-|101
\rangle) \to -\beta|1
\rangle_B$
    * $\implies |10
\rangle_{CA} \otimes (\alpha|0
\rangle - \beta|1
\rangle)_B$

* **$|11
\rangle_{CA}$:** (Find all terms that start with $|11
\rangle$)
    * $\alpha|111
\rangle \to \alpha|1
\rangle_B$
    * $\beta(-|110
\rangle) \to -\beta|0
\rangle_B$
    * $\implies |11
\rangle_{CA} \otimes (-\beta|0
\rangle + \alpha|1
\rangle)_B$

This regrouped state is the key to the whole protocol!

--- 

## 5. Alice's Measurement and Bob's Correction

Alice now measures her two qubits, $q_C$ and $q_A$. Her measurement (a "Z-basis measurement") will be $00$, $01$, $10$, or $11$, each with a 25% probability. This measurement *collapses* the entire 3-qubit state into one of the four terms we just grouped.

Bob's qubit, $q_B$, is now in a new state, *conditional* on Alice's measurement.

**Case 1: Alice measures $00$**
* The system collapses to: $\frac{1}{2} |00
\rangle_{CA} \otimes (\alpha|0
\rangle + \beta|1
\rangle)_B$
* Bob's qubit is: $(\alpha|0
\rangle + \beta|1
\rangle)$. **This is exactly $|\psi
\rangle$!**
* Alice sends "00" to Bob. Bob does nothing (applies an $I$ gate).

**Case 2: Alice measures $01$**
* The system collapses to: $\frac{1}{2} |01
\rangle_{CA} \otimes (\beta|0
\rangle + \alpha|1
\rangle)_B$
* Bob's qubit is: $(\beta|0
\rangle + \alpha|1
\rangle)$. This is $|\psi
\rangle$ but with the amplitudes swapped.
* Alice sends "01" to Bob. Bob applies a **Pauli-$X$ gate**:
    * $X (\beta|0
\rangle + \alpha|1
\rangle) = \beta|1
\rangle + \alpha|0
\rangle = \alpha|0
\rangle + \beta|1
\rangle = |\psi
\rangle$.
    * This Pauli-$X$ gate protocol ensure that the information of $|\psi\rangle$ is transferred exactly to Bob

**Case 3: Alice measures $10$**
* The system collapses to: $\frac{1}{2} |10
\rangle_{CA} \otimes (\alpha|0
\rangle - \beta|1
\rangle)_B$
* Bob's qubit is: $(\alpha|0
\rangle - \beta|1
\rangle)$. This is $|\psi
\rangle$ but with a phase flip on $|1
\rangle$.
* Alice sends "10" to Bob. Bob applies a **Pauli-$Z$ gate**:
    * $Z (\alpha|0
\rangle - \beta|1
\rangle) = \alpha|0
\rangle - (-\beta|1
\rangle) = \alpha|0
\rangle + \beta|1
\rangle = |\psi
\rangle$.

**Case 4: Alice measures $11$**
* The system collapses to: $\frac{1}{2} |11
\rangle_{CA} \otimes (-\beta|0
\rangle + \alpha|1
\rangle)_B$
* Bob's qubit is: $(-\beta|0
\rangle + \alpha|1
\rangle)$. This state is both swapped *and* phase-flipped.
* Alice sends "11" to Bob. Bob applies an **$X$ gate, then a $Z$ gate**:
    * **First $X$:** $X(-\beta|0
\rangle + \alpha|1
\rangle) = -\beta|1
\rangle + \alpha|0
\rangle$
    * **Then $Z$:** $Z(\alpha|0
\rangle - \beta|1
\rangle) = \alpha|0
\rangle - (-\beta|1
\rangle) = \alpha|0
\rangle + \beta|1
\rangle = |\psi
\rangle$.

--- 

## 6. Conclusion

In all four possible outcomes, Bob is able to perfectly reconstruct the original state $|\psi
\rangle$ by applying a specific correction (I, X, Z, or ZX) based on the two classical bits he receives from Alice.

The protocol successfully "teleported" the unknown state $|\psi
\rangle$ from Alice to Bob, transefering the information(qubit state) to Bob. This protocol does not limit by the distance between Alice and Bob, the qubit state collapsed immediately on the Bob side when Alice measured and collapse her own qubit($q_A$). However, the information does not travelled instantaneously, as Alice has to tell Bob what gate to apply, thus the information does not travel at or faster than light speed.

In the process, Alice's original state was destroyed by her measurement, satisfying the No-Cloning Theorem. This derivation shows how entanglement serves as a quantum resource, activated by classical communication, to achieve a task impossible in classical physics.
