### Question 1

What is the *primary resource* that makes quantum teleportation possible, which Alice and Bob must share *before* the protocol begins?

### Question 2

In the protocol, Alice has two qubits: the 'cat' qubit ($q_C$) and her half of the Bell pair ($q_A$). What specific operation does she perform on these two qubits?

A) She measures them both in the Z-basis immediately.

B) She performs a Bell Basis Measurement (CNOT, then H, then measure).

C) She applies an $X$ gate to $q_C$ and a $Z$ gate to $q_A$.

D) She sends both qubits to Bob through a quantum channel.

**Answer & Explanation:**

**B)** is correct. Alice must measure her two qubits in the Bell basis. The circuit to do this is to first apply a $CNOT(C, A)$, then a $H(C)$, and *then* measure both $q_C$ and $q_A$ in the Z-basis.

### Question 3

What information does Alice send to Bob to complete the teleportation?

C) Her half of the entangled pair, $q_A$.

**Answer & Explanation:**

**B)** is correct. Alice measures her two qubits, which collapses them into one of four classical outcomes (00, 01, 10, or 11). She must send this two-bit classical result to Bob so he knows which correction gate to apply.

### Question 4

**B)** is correct. The protocol *moves* a state, it doesn't *copy* it. Alice's measurement on $q_C$ and $q_A$ inevitably destroys the original $|\psi

### Question 5

Why does quantum teleportation *not* allow for faster-than-light (FTL) communication?

A) The entanglement resource ($|\Phi^+\rangle$) must be physically mailed from Alice to Bob first.

B) The final correction step requires Alice to send two classical bits to Bob, and this message is limited by the speed of light.

C) The CNOT and H gates are very slow to perform, adding significant delay.

D) It *does* allow FTL communication, which is why it's so revolutionary.

**Answer & Explanation:**

**B)** is correct. This is a crucial point. While Alice's measurement *instantly* collapses Bob's qubit into one of four states, Bob has no idea *which* state it is (e.g., $|\psi
\rangle$, $X|\psi
\rangle$, $Z|\psi
\rangle$, or $ZX|\psi
\rangle$) until he receives Alice's classical instructions. That classical message cannot travel faster than light.
