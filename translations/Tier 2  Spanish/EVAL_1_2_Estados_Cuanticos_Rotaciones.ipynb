> **Prop√≥sito:** Vincular el razonamiento algebraico con la interpretaci√≥n geom√©trica.

**Concepto clave:**

$$

> **Interpretaci√≥n:** La rotaci√≥n cambia la *orientaci√≥n* del estado en la esfera de Bloch, pero mantiene su probabilidad total igual a 1.
